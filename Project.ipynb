```python
# count the total number of bookings and cancellations for each property
counts = data_booking.groupby('property_name')['booking_status'].value_counts().unstack().fillna(0)

# calculate the cancellation rate for each property
cancellation_rate = ((counts['Cancelled'] / counts.sum(axis=1))*100).reset_index()

# rename the columns and sort by cancellation rate in descending order
cancellation_rate.columns = ['property_name', 'cancellation_rate']
cancellation_rate = cancellation_rate.sort_values('cancellation_rate', ascending=False).reset_index(drop=True)

# show the result
print(cancellation_rate)
```

```python
# Function to extract the month using startswith()
def extract_month(booking_id):
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    for month in months:
        if booking_id.startswith(month):
            return month
    return None

# Apply the function to the 'Booking ID' column to extract the months
data_booking['Month'] = data_booking['booking_id'].apply(extract_month)
data_booking
```
